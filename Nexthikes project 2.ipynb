```python
from scipy import stats
```

```python
Final_dataset['Z_Score'] = stats.zscore(Final_dataset['cnt'])
Final_dataset['Outlier_z'] = (Final_dataset['Z_Score'].abs() >3)
Final_dataset
```

```python
z = np.abs(stats.zscore(Final_dataset['cnt']))
print(z)
```

```python
# To define an outlier threshold value is chosen which is generally 3.0. 
# As 99.7% of the data points lie between +/- 3 standard deviation, so we take threshold as 2
threshold_z = 2
 
outlier_indices = np.where(z > threshold_z)[0]
no_outliers = Final_dataset.drop(outlier_indices)
print("Original DataFrame Shape:", Final_dataset.shape)
print("DataFrame Shape after Removing Outliers:", no_outliers.shape)
```

IQR identifies outliers as data points falling outside the range defined by Q1-k*(Q3-Q1) and Q3+k*(Q3-Q1), where Q1 and Q3 are the first and third quartiles, and k is a factor (typically 1.5).
so here
IQR => from (-84.5) to (+183.5)
Q1-> 16
Q3-> 83
k-> 1.5
Outliers are present

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def removal_box_plot(df, column, threshold):
	sns.boxplot(df[column])
	plt.title(f'Original Box Plot of {column}')
	plt.show()

	removed_outliers = df[df[column] <= threshold]

	sns.boxplot(removed_outliers[column])
	plt.title(f'Box Plot without Outliers of {column}')
	plt.show()
	return removed_outliers


threshold_value = +183.5

no_outliers = removal_box_plot(Final_dataset, 'cnt', threshold_value)
```
